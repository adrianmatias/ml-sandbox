```python
from src.const import LLM
from src.evaluation.eval_set import EvalSet

model_aug_list = [
    LLM.QWEN_3_5_27B_Q2,
    LLM.QWEN_3_5_9B,
    LLM.GPT_OSS_20B,
]

eval_set = EvalSet()
eval_set.load()
```

```python
from run.evaluate import evaluate_model

all_results = {}
for aug in model_aug_list:
    print(f"Evaluating {aug}...")
    all_results[aug] = await evaluate_model(aug=aug, eval_set=eval_set)
```

```python
import pandas as pd

from src.const import CONST
```

```python
def get_report_df(all_results) -> pd.DataFrame:

    metrics = CONST.eval.metric_list
    rows = []

    for aug, results in all_results.items():
        row = {"model": aug}
        for m in metrics:
            scores = [r[m] for r in results if isinstance(r.get(m), (int, float))]
            row[m] = sum(scores) / len(scores) if scores else 0.0
        rows.append(row)
    df = pd.DataFrame(rows).set_index("model")

    return df


report_df = get_report_df(all_results=all_results)
report_df
```
